This notebook explores visualization plots in matplotlib library using stocks data: Line plots, Line plots with subplot, Scatter Plot, and Histogram Plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the stock prices data csv file using pandas
stock_df = pd.read_csv('stocks.csv')
stock_df[0:5]
```

```python
# Line Plot using Matplotlib
stock_df.plot(x = 'Date', y = ['AAPL', 'TSLA','T'], label = ['AAPL', 'TSLA','T'], linewidth = 2, colormap='winter');
plt.ylabel('Price of Stock')
plt.title('Stocks')
plt.grid()
```

```python
# Subplots of Line Plot using Matplotlib

plt.figure(figsize = (6,7))
plt.subplot(3, 1, 1)
plt.plot(stock_df.index, stock_df['AAPL'], 'r--', linewidth = 3);
plt.grid()
plt.legend(['APPLE price'])

plt.subplot(3, 1, 2)
plt.plot(stock_df.index, stock_df['TSLA'], 'b.', linewidth = 1);
plt.grid()
plt.legend(['TSLA price'])


plt.subplot(3, 1, 3)
plt.plot(stock_df.index, stock_df['GOOG'], 'g-');
plt.grid()
plt.legend(['GOOGLE price'])
```

```python
# Read the daily_returns data csv file using pandas
daily_return_df = pd.read_csv('daily_returns.csv')
daily_return_df[0:5]
```

```python
# Scatter Plot using Matplotlib

X = daily_return_df['AAPL']
Y = daily_return_df['sp500']

plt.scatter(X, Y);
plt.title('Daily Return')
plt.ylabel('AAPL')
plt.xlabel('sp500')
```

```python
# Histogram of the daily returns

mu1 = daily_return_df['AAPL'].mean() # mean 
sigma1 = daily_return_df['AAPL'].std() # std dev 
mu1 = (round(mu1, 5));
sigma1 = (round(sigma1, 5));

mu2 = daily_return_df['GOOG'].mean() # mean 
sigma2 = daily_return_df['GOOG'].std() # std dev 
mu2 = (round(mu2, 5));
sigma2 = (round(sigma2, 5));
num_bins = 50

plt.figure(figsize = (6, 5)) 
plt.subplot(2, 1, 1)
plt.hist(daily_return_df['AAPL'], num_bins, facecolor = 'green');
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu1) + ', sigma=' +str(sigma1));

plt.figure(figsize = (6, 5)) 
plt.subplot(2, 1, 2)
plt.hist(daily_return_df['GOOG'], num_bins, facecolor = 'red');
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu2) + ', sigma=' +str(sigma2));
```
